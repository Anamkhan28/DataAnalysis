```python
import pandas
import seaborn
from matplotlib import pyplot
```

```python
data={
    'Order_id':[1001,1002,1003,1004,1005,1006],
    'Customer_name':['Ravi','Ananya','Sahil','Divya','Amit','Divya'],
    'Item':['Burger','Maggi','Dosa','Pizza','Pasta','Momos'],
    'Price':[150,100,120,300,210,100],
    'Quantity':[1,2,1,2,3,2],
    'Payment_mode':['UPI','Cash','Card','UPI','Cash','Cash']
}
```

```python
df=pandas.DataFrame(data)
print(df)
```

```python
seaborn.lineplot(x='Customer_name',y='Quantity',data=df)
pyplot.title('Restaurant Orders')
pyplot.show()
```

```python
seaborn.countplot(x="Payment_mode", data=df)
pyplot.title("Orders by Payment Mode")
pyplot.show()
```

```python
seaborn.barplot(x='Price',y='Item',data=df)
pyplot.title('Restaurant Orders')
pyplot.show()
```

```python
seaborn.barplot(x='Customer_name',y='Quantity',data=df)
pyplot.title('Restaurant Orders')
pyplot.show()
```

```python
seaborn.lineplot(x='Customer_name',y='Price',data=df)
pyplot.title('Restaurant Orders')
pyplot.show()
```

```python
seaborn.scatterplot(x='Quantity',y='Item',data=df)
pyplot.title('E-Commerce Product Sales')
pyplot.show()
```
