```python
import pandas
import matplotlib.pyplot as plt
```

```python
data={
    'Brand':['Samsung','Apple','OnePlus','Vivo','Nokia','Motorola'],
    'Model':['Galaxy M14','Iphone 16','Nord CE','Vivo T2','Nokia G42','Moto G73'],
    'Price':[13500,62900,19000,16000,12500,18000],
    'RAM':[6,8,6,4,8,6],
    'Rating': [4.3, 4.2, 4.7, 2.1, 3.4, 2.0]
}
```

```python
df=pandas.DataFrame(data)
print(df)
```

```python
plt.plot(df['Brand'],df['Price'])
plt.title('Price of Mobile')
plt.xlabel('Brand')
plt.ylabel('Price')
plt.grid(True)
plt.show()
```

```python
plt.bar(df['Brand'],df['Rating'])
plt.title('Rating of Mobile')
plt.xlabel('Brand')
plt.ylabel('Rating')
plt.show()
```

```python
plt.bar(df['Model'],df['RAM'],color='Pink')
plt.title('Model of mobiles')
plt.xlabel('Brand')
plt.ylabel('RAM of mobiles')
plt.show()
```

```python
plt.barh(df['Model'],df['RAM'],color='skyblue')
plt.title('Model of mobiles')
plt.ylabel('Brand')
plt.xlabel('RAM of mobiles')
plt.show()
```

```python
plt.scatter(df['Model'],df['Price'],color='green')
plt.plot(df['Model'],df['Price'],color='brown')
plt.title('Model of mobiles')
plt.xlabel('Model')
plt.ylabel('Price')
plt.show()
```

```python
sizes=df['Rating']
labels=df['Price']

plt.pie(sizes,labels=labels,autopct='%1.1f%%')
plt.title('Rating of Mobiles')
plt.show()
```
