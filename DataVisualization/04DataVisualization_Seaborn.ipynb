```python
import pandas
import seaborn
from matplotlib import pyplot
```

```python
data={
    'Product':['Wireless Mouse','Laptop','Smartwatch','Keyboard','Hard Drive','Power Bank'],
    'Price':[999,55000,3999,1799,3499,1199],
    'Units_Sold': [150, 85, 200, 120, 30, 60,],
    'Category': ["Accessories","Computers","Wearables","Accessories","Storage","Accessories"],
    'Rating': [4.3, 4.5, 4.0, 4.2, 4.7, 4.4]

}
```

```python
df=pandas.DataFrame(data)
print(df)
```

```python
seaborn.lineplot(x='Product',y='Price',data=df)
pyplot.title('E-Commerce Product Sales')
pyplot.show()
```

```python
seaborn.barplot(x='Product',y='Price',data=df)
pyplot.title('E-Commerce Product Sales')
pyplot.show()
```

```python
seaborn.barplot(x='Price',y='Product',data=df)
pyplot.title('E-Commerce Product Sales')
pyplot.show()
```

```python
seaborn.scatterplot(x='Product',y='Price',data=df)
pyplot.title('E-Commerce Product Sales')
pyplot.show()
```

```python
seaborn.histplot(df['Price'])
pyplot.title('E-Commerce Product Sales')
pyplot.show()
```

```python
seaborn.barplot(x="Product", y="units_sold", data=df)
pyplot.xticks(rotation=45)
pyplot.title("Units Sold per Product")
pyplot.show()
```

```python
seaborn.boxplot(x="Category", y="Rating", data=df)
pyplot.title("Customer Ratings by Category")
pyplot.show()
```

```python
sizes=df['Units_Sold']
labels=df['Price']

pyplot.pie(sizes,labels=labels,autopct='%1.1f%%')
pyplot.title('Products ')
pyplot.show()
```
